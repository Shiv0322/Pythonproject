```python
#visualizing the class distribution for target variable(our target variable is placement status)

sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

placement count of male is greater than placement count of female.

```python
#Understand how many students got placed vs not placed.
sns.countplot(x='status', data=df)
plt.title('Placement Status Count')
plt.show()
```

more students are  placed in campus placement.

```python
sns.barplot(x='degree_t', y='salary', data=df)
plt.title('Average Salary by Degree Type')
plt.xticks(rotation=45)
plt.show()
```

The person who have trade sci and tech have more average salary than commerce and others.


See which undergraduate degree type results in higher average salary.



```python
sns.boxplot(x='gender', y='ssc_p', data=df)
plt.title('gender vs ssc_p')
plt.show()
```

ssc percent of females are more than male




```python
sns.scatterplot(x='mba_p', y='salary', hue='status', data=df)
plt.title('MBA Percentage vs Salary')
plt.show()
```

most of the students having the percent between 55 to 70 placed with salary 200000 to 300000

```python
#correlation matrix
num_cols=['ssc_p','salary','hsc_p','status']
sns.pairplot(df[num_cols],diag_kind='kde')
plt.suptitle('pairplot of numerical features',y=1.02)
plt.show()
```
